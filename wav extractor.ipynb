```python
X = [] # will hold PSDs of the recs
X_names = [] # will hold the full path and the filename of all recs
y = [] # the labels
target_count = []

for i, target in enumerate(target_names):
    target_count.append(0)
    #path = './Wingbeats/' + target + '/'
    path = folder_path + target + '\\'
    for [root, dirs, files] in os.walk(path, topdown = False):
        for filename in files:
            name,ext = os.path.splitext(filename)
            if ext == '.wav':
                name = os.path.join(root, filename)
                data, fs = librosa.load(name, sr = 8000)
                y.append(i)
                X.append([10*np.log10(signal.welch(data, fs=fs, window='hanning', nperseg=256, noverlap=128+64)[1])])
                X_names.append(name)
                target_count[i]+=1

    print (target, '#recs = ', target_count[i])

print ('total #recs = ', len(y))
```

```python
species = []
for entry in X_names:
    if "pipiens" in entry:
        species.append(target_names[1])
    else:
        species.append(target_names[0])
```

```python
time = []
#timespot = re.search(r'\\D_(.+?)\\F')
for entry in X_names:
    thing = entry.partition(r'\D_')[2].partition(r'\F')
    timestring = thing[0].split('_')
    hour = timestring[3]
    time.append(hour)
print(time[0:3])
```

```python
names = pd.DataFrame(X_names, columns=['name'])
PSDs = pd.DataFrame(X, columns=['PSD'])
frame = pd.concat([names, PSDs], axis = 1)
frame['filename'] = frame['name'].str.extract('([F]\w{0,})',expand=True)
frame['species'] = pd.DataFrame(species)
frame['hour'] = pd.DataFrame(time)


frame.head()
```
